```python
# pyexcel_xls wrapper library to read, manipulate and write data in xls format
from pyexcel_xls import save_data
from pyexcel_xls import read_data
#urllib is a package that collects several modules for working with URLs
#request is used for making as request
import urllib.request
#Beautiful Soup is a Python package for parsing HTML and XML documents
from bs4 import BeautifulSoup
#XlsxWriter is a Python module for creating Excel XLSX files.
#XlsxWriter is a Python module that can be used to write text, numbers, 
import xlsxwriter
#JSON (JavaScript Object Notation) is a lightweight data-interchange format.
import json

data=read_data("File1.xls")
# dump and load convert between files and objects,
#while dumps and loads convert between strings and objects
s1=json.dumps(data)
object = json.loads(s1)
urlNames=dict()
 #Workbook class is the main class exposed by the XlsxWriter module
wb=xlsxwriter.Workbook('File2.xlsx')  
f1=wb.add_format({'bold':True,'font_color':'yellow'})
f2=wb.add_format({'bold':True,'font_color':''})
f3=wb.add_format({'bold':True,'font_color':''})



for o in object:
    ws=wb.add_worksheet(str(o))
    ws.write("A2",object[o][0][0],f1)
    ws.write("A4","",f2)
    ws.write("B5","",f3)
    print(f1)
    # Create a new Chart object.
    chart=wb.add_chart({'type':'column'})
    chart1 = wb.add_chart({'type': 'line'})
   # chart2 = wb.add_chart({'type': 'pie'})
    
    req=urllib.request.Request(str(object[o][0][0]),data=None,headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'})
    f=urllib.request.urlopen(req)
    s=f.read().decode('utf-8')
    soup=BeautifulSoup(s,"html.parser")
    for script in soup(["script","style"]):
        script.extract()
    t=soup.get_text()
    text = "".join([s for s in t.splitlines(True) if s.strip("\r\n")])
    #Important line to split to catch the words
    print("Split",text)
    L=text.split()
    wor=[]
    for i in range(1,len(object[o])):
        wor.append(object[o][i][0])
    #To count Total words in link    
    print("Inside for loop",i)    
    print("WordTobeSorted",wor)#File1ExcelWordsSearching       
    words=sorted(set(wor))#To Sorting File1ExcelData
    print("WordsSorted",words)
    countD=dict()
    i=3
    for w in words:
       countD[w]=L.count(w)
       ws.write(i,0,w,f3)
       ws.write(i,1,L.count(w),f3)
       i+=1
       #View All words 
       print("ChartForloop",w)    
    area="="+str(o)+"!$B$4:$B$"+str(i)     
    cate="="+str(o)+"!$A$4:$A$"+str(i)
    #print(area,cate)
    
    #Creating chart1
    chart.add_series({'categories':cate ,'values': area})
    
    chart.set_title({'name': 'WORD COUNT'})
    chart.set_x_axis({'name': 'Words'})
    chart.set_y_axis({'name': 'Count','num_font':  {'rotation': 45}})
    
    chart.set_style(37)
    #ChartPosition to be insert
    ws.insert_chart("J4",chart)

     
    #Creating another chart2
    chart1.add_series({'categories':cate ,'values': area})
    
    chart1.set_title({'name': 'WORD COUNT'})
    chart1.set_x_axis({'name': 'Words'})
    chart1.set_y_axis({'name': 'Count','num_font':  {'rotation': 45}})
    
    chart1.set_style(37)
    #ChartPosition to be insert
    ws.insert_chart("J22",chart1)


    
    #Display in PythonShell
    print(countD)
    
    print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
    
wb.close()






```
